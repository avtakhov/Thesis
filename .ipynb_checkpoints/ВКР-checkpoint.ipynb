Подход основан на том, чтобы оценивать компанию в раннем этапе ее развития. Ранний этап развития ограничим рамками в 4 года со времени основания компании. Таким образом, мы будем собирать выводить признаки, лишь касающиеся первых 4-лет развития компании.

Метки, определяющие успех/провал компании:
- если компания не получает никакого инвестирования спустя 4 года после основания, то компания **провальная**
- если компания находится в статусе "закрытой", но не будучи поглощенной, то компания **провальная**
- если компания достигает следующего раунда инвестирования в течение 3-х лет после прошествия раннего этапа развитии компании, то компания **успешная**
- если компания выходит на IPO, то компания **успешная**
- если компания поглощается другой, то компания **успешная**

Компании, которые будут исключены из анализа:
- если компания в течении раннего этапа развития достигает раунда C
- если компания в течении раннего этапа развития выходит на IPO
- если компания в течении раннего этапа развития поглощена
- если компания создана с 2010 года (включительно)

В исследовании будут участвовать компании, созданные с 2006 по 2009, чтобы максимально актуализировать исследование. 

# 3. Предобработка данных

Прежде всего, выберем компании, которые будут участвовать в исследовании:

```python
def delete_from(selected_companies: list, companies_to_delete : list) -> list:
    values_set = set(selected_companies)
    for c in companies_to_delete:
        if c in values_set:
            values_set.remove(c)
    return list(values_set)

#выберем компании с 1997 по 2009 года
objects.founded_at = pd.to_datetime(objects.founded_at, format = '%Y-%m-%d')
objects.founded_at = pd.DatetimeIndex(objects.founded_at).year

objects = objects[objects.entity_type == 'Company'][objects.founded_at < 2009][objects.founded_at > 1997]
selected_companies = objects.id.values.tolist()

#выберем компании, которые не прошли ipo в течение 4-х лет с момента основания
ipos = ipos[ipos.object_id.isin(selected_companies)]
ipos.public_at = pd.to_datetime(ipos.public_at, format = '%Y-%m-%d')
ipos.public_at = pd.DatetimeIndex(ipos.public_at).year
ipos = ipos.merge(objects, how = 'inner', left_on = 'object_id', right_on = 'id')
companies_to_delete = ipos.loc[(ipos.public_at - ipos.founded_at) <= 4].object_id.values.tolist()

#выберем компании, которые не были поглощены в течение 4-х лет с момента основания
acquisitions = acquisitions[acquisitions.acquired_object_id.isin(selected_companies)]
acquisitions.acquired_at = pd.to_datetime(acquisitions.acquired_at, format = '%Y-%m-%d')
acquisitions.acquired_at = pd.DatetimeIndex(acquisitions.acquired_at).year
acquisitions = acquisitions.merge(objects, how = 'inner', left_on = 'acquired_object_id', right_on = 'id')
companies_to_delete += acquisitions.loc[(acquisitions.acquired_at - acquisitions.founded_at) <= 4].acquired_object_id.values.tolist()
 
#выберем компании, которые не достигли раунда C и выше в течение 4-х лет с момента основания
funding_rounds = funding_rounds[funding_rounds.object_id.isin(selected_companies)]
funding_rounds = funding_rounds[funding_rounds.funding_round_type == 'series-c+']
funding_rounds.funded_at = pd.to_datetime(funding_rounds.funded_at, format = '%Y-%m-%d')
funding_rounds.funded_at = pd.DatetimeIndex(funding_rounds.funded_at).year
funding_rounds = funding_rounds.merge(objects, how = 'inner', left_on = 'object_id', right_on = 'id')
companies_to_delete += funding_rounds.loc[(funding_rounds.funded_at - funding_rounds.founded_at) <= 4].object_id.values.tolist()

selected_companies = delete_from(selected_companies, companies_to_delete)
```

### 3.1.3. Преобразование данных

Преобразуем дату раунда инвестирования в год и выделим этот год в качестве отдельного признака:

```python
funding_rounds['funded_at'] = pd.to_datetime(funding_rounds['funded_at'], 
                                             format = '%Y%m%d', infer_datetime_format=True)
funding_rounds['funded_at_year'] = pd.DatetimeIndex(funding_rounds['funded_at']).year
```

Добавим год создания компаний, чтобы ограничить ранний этап развития компании:

```python
early_stage = funding_rounds.merge(objects[['id', 'founded_at']], how = 'inner', left_on = 'object_id', right_on = 'id')
early_stage.founded_at = pd.to_datetime(early_stage.founded_at, format = '%Y-%m-%d')
early_stage['founded_at_year'] = pd.DatetimeIndex(early_stage.founded_at).year
early_stage = early_stage.loc[(early_stage.funded_at_year - early_stage.founded_at_year) <= 4]
early_stage.head()
```

Введем новые признаки: **funding_rounds, first_round_amount, first_round_participants, first_round_type, second_round_amount, second_round_participants, second_round_type, last_round_amount, last_round_participants, last_round_type, has_debt_round**.

1) **funding_rounds**:

```python
rounds_number_df = pd.DataFrame(columns = ['company', 'rounds_number'])

for row, company in enumerate(set(early_stage.object_id.values)):
    rounds_number_df.loc[row] = [company, len(early_stage[early_stage.object_id == company])]

rounds_number_df.head()
```

2) **first_round_amount, first_round_participants, first_round_type**

```python
first_round_df = pd.DataFrame(columns = ['company', 'first_round_amount', 'first_round_participants', 'first_round_type'])

for row, company in enumerate(set(early_stage.object_id.values)):
    first_round = early_stage[early_stage.object_id == company].sort_values(by = 'funded_at').iloc[0]
    first_round_amount = first_round.raised_amount_usd
    first_round_participants = first_round.participants
    first_round_type = first_round.funding_round_type
    first_round_df.loc[row] = [company, first_round_amount, first_round_participants, first_round_type]

first_round_df.head()
```

3) **second_round_amount, second_round_participants, second_round_type**:

```python
second_round_df = pd.DataFrame(columns = ['company', 'second_round_amount', 'second_round_participants', 'second_round_type'])

for row, company in enumerate(set(early_stage.object_id.values)):
    if len(early_stage[early_stage.object_id == company]) > 1:
        second_round = early_stage[early_stage.object_id == company].sort_values(by = 'funded_at').iloc[1]
        second_round_amount = second_round.raised_amount_usd
        second_round_participants = second_round.participants
        second_round_type = second_round.funding_round_type
        second_round_df.loc[row] = [company, second_round_amount, second_round_participants, second_round_type]

second_round_df.head()
```

4) **last_round_amount, last_round_participants, last_round_type**:

```python
last_round_df = pd.DataFrame(columns = ['company', 'last_round_amount', 'last_round_participants', 'last_round_type'])

for row, company in enumerate(set(early_stage.object_id.values)):
    last_round = early_stage[early_stage.object_id == company].sort_values(by = 'funded_at').iloc[-1]
    last_round_amount = last_round.raised_amount_usd
    last_round_participants = last_round.participants
    last_round_type = last_round.funding_round_type
    last_round_df.loc[row] = [company, last_round_amount, last_round_participants, last_round_type]

last_round_df.head()
```

5) **has_debt_round**:

```python
has_debt_round_df = pd.DataFrame(columns = ['company', 'has_debt_round'])

for row, company in enumerate(set(early_stage.object_id.values)):
    if 'debt_round' in early_stage[early_stage.object_id == company].funding_round_code.values:
        has_debt_round_df.loc[row] = [company, 1]
    else:
        has_debt_round_df.loc[row] = [company, 0]

has_debt_round_df.head()
```

### 3.1.4. Объединение таблиц в единую

Все по отдельности созданные выше таблицы объединим в одну:

```python
funding_rounds_final = early_stage.groupby('id')['raised_amount_usd', 'participants'].sum().reset_index()
funding_rounds_final.columns = ['company', 'raised_amount_usd', 'participants']
funding_rounds_final = funding_rounds_final.merge(rounds_number_df, how = 'left', on = 'company').\
                                            merge(first_round_df, how = 'left', on = 'company').\
                                            merge(second_round_df, how = 'left', on = 'company').\
                                            merge(last_round_df, how = 'left', on = 'company').\
                                            merge(has_debt_round_df, how = 'left', on = 'company')

funding_rounds_final.head()
```

Заполним пропуски нулями или 'none' (в случае с пропущенным типом раунда инвестирования):

```python
funding_rounds_final.isnull().sum()
```

```python
funding_rounds_final.second_round_amount = funding_rounds_final.second_round_amount.fillna(0)
funding_rounds_final.second_round_participants = funding_rounds_final.second_round_participants.fillna(0)
funding_rounds_final.second_round_type = funding_rounds_final.second_round_type.fillna('none')
```

Как итог, 90% данных – это провальные компании, 10% – успешные. Чтобы устранить дисбаланс классом, применим метод ADASYN, который добавит наблюдений к минорному классу:

```python
X = objects_final.drop('success', axis = 1)
y = objects_final['success']

sm = ADASYN()
X, y = sm.fit_resample(X, y)

objects_final = X
objects_final['success'] = y
```

#### 4.4.3. Изучим зависимость признаков от целевого признака

Проверять зависимость мы будем через корреляцию Пирсона и p-значение:

```python
cols = list(objects_final.columns)

for col in cols: 
    pearson_coef, p_value = stats.pearsonr(objects_final[col], objects_final['success'])
    print(f'PearsonR между {col} и success {pearson_coef} с P-знчением P = {p_value}')
    
    if p_value < 0.001:
        print(f'Корреляция между {col} и success статистически значимая')
    elif p_value < 0.05:
        print(f'Корреляция между {col} и success средняя')
    elif p_value < 0.1:
        print(f'Корреляция между {col} и success слабая')
    else:
        print(f'Корреляция между {col} и success статистически незначимая')
        
    if pearson_coef > 0 :
        if pearson_coef > 0.85:
            print(f'Коэффициент ~{pearson_coef} показывает положительную очень сильную связь\n')
        elif pearson_coef > 0.75 :
            print(f'Коэффициент ~{pearson_coef} показывает положительную достаточно сильную связь\n')
        elif pearson_coef > 0.60:
            print(f'Коэффициент ~{pearson_coef} показывает положительную относительно сильную связь\n')
        elif pearson_coef > 0.50 :
            print(f'Коэффициент ~{pearson_coef} показывает положительную среднюю связь\n')
        else:
            print(f'Коэффициент ~{pearson_coef} показывает положительную слабую связь\n')
    else:
        if abs(pearson_coef) > 0.85:
            print(f'Коэффициент ~{pearson_coef} показывает негативную очень сильную связь\n')
        elif abs(pearson_coef) > 0.75 :
            print(f'Коэффициент ~{} показывает негативную достаточно сильную связь\n')
        elif abs(pearson_coef) > 0.60:
            print(f'Коэффициент ~{pearson_coef} показывает негативную относительно сильную связь\n')
        elif abs(pearson_coef) > 0.50 :
            print(f'Коэффициент ~{pearson_coef} показывает положительную среднюю связь\n')
        else:
            print(f'Коэффициент ~{pearson_coef} показывает положительную слабую связь\n')
```

Если связь с целевой переменной статистически незначимая, то от переменной можно избавиться. Пропустим этот ставим этот шаг и вернемся к нему в случае, если модель покажет неудовлетворительный результат.

### 4.5. Вводим модели машинного обучения

```python
#удалим лишние столбцы
objects_final.drop(['ipo', 'acquired'], axis = 1, inplace = True)
```

```python
X = objects_final.drop('success', axis = 1)
y = objects_final['success']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
models = []
models.append(('LR', LogisticRegression()))
models.append(('LDA', LinearDiscriminantAnalysis()))
models.append(('KNN', KNeighborsClassifier()))
models.append(('CART', DecisionTreeClassifier()))
models.append(('NB', GaussianNB()))
models.append(('RF', RandomForestClassifier()))

results = []
names = []
scoring = 'accuracy'

for name, model in models:
    kfold = KFold(n_splits=5)
    cv_results = cross_val_score(model, X_train, y_train, cv=kfold, scoring=scoring)
    results.append(cv_results)
    names.append(name)
    msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
    print(msg)
```

Лучший результат показала модель случайного леса, 94%, что является отличным результатом.

#### Протестируем модель на тестовых данных

```python
model = RandomForestClassifier()
model.fit(X_train, y_train)
```

Проверим качество модели на 3-х метриках:

```python
print(f'Precision - {precision_score(y_test, model.predict(X_test))}')
print(f'Recall - {recall_score(y_test, model.predict(X_test))}')
print(f'Accuracy - {accuracy_score(y_test, model.predict(X_test))}')
```
